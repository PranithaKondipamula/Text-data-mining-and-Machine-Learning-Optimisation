```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('relevant_model_outputs.xlsx')
```

```python
df.info()
```

```python
plt.bar(df['Model'], df['Test Upper CI'], color = 'y')
plt.bar(df['Model'], df['Test Lower CI'], color = 'r')
plt.ylim([0.9,0.94])
plt.ylabel('Accuracy CI')
plt.xlabel('Model')
plt.title('Original Categories - Best Models by CI')
plt.xticks(rotation=-90)
plt.figure(figsize=(20,20))
plt.show()
```

```python
plt.scatter(range(len(df)), df['Test Accuracy'],color='blue')
for lower,upper,y in zip(df['Test Lower CI'],df['Test Upper CI'],range(len(df))):
    plt.plot((y,y),(lower,upper),'ro-',color='red')
plt.xticks(ticks = range(len(df)),labels = list(df['Model']),rotation=45, ha='right', rotation_mode='anchor')
plt.title('Accuracy C.I.s for the Top 21 Models')
plt.ylabel('Accuracy')
plt.xlabel('Model')
plt.show()
```
